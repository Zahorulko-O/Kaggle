```python
%%time
# read the data and fill the missing values
path='catch-me-if-you-can-intruder-detection-through-webpage-session-tracking2'
times = ['time%s' % i for i in range(1, 11)]
sites = ['site%s' % i for i in range(1, 11)]
train_df = pd.read_csv(os.path.join(path, 'train_sessions.csv'), index_col='session_id')
test_df = pd.read_csv(os.path.join(path, 'test_sessions.csv'), index_col='session_id')

y = train_df['target']
train_df.fillna(0, inplace = True)
test_df.fillna(0, inplace = True)
test_df[times] = test_df[times].astype('datetime64')
train_df[times] = train_df[times].astype('datetime64')
y.astype('int8').values
test_df
```

```python
# create train/test datas
test_sessions =  test_df.loc[:,(sites)].copy()
train_sessions = train_df.loc[:,(sites)].copy()
test_times =  test_df.loc[:,(times)].copy()
train_times = train_df.loc[:,(times)].copy()
train_sessions
```

```python
%%time 
def prepare_sparse_data(df_csv, dic_pkl):
    df_csv[sites] = df_csv[sites]
   # df_csv[times] = df_csv[times].astype('datetime64')
    df_csv.fillna(0, inplace = True)
    with open(dic_pkl, "rb") as input_dict:
        site_dict=pickle.load(input_dict)
    name_site = {val:key for (key, val) in site_dict.items()}
    name_site[0] = 'unknown'
    df_csv = df_csv[sites].apply(lambda row: ' '.join([name_site[i] for i in row]), axis=1).tolist()
    
    return df_csv

test_sessions  = prepare_sparse_data(test_sessions,  
                      "catch-me-if-you-can-intruder-detection-through-webpage-session-tracking2/site_dic.pkl")  
train_sessions = prepare_sparse_data(train_sessions,
                         "catch-me-if-you-can-intruder-detection-through-webpage-session-tracking2/site_dic.pkl") 
```

```python
%%time
#transform test/train sessions data
vectorizer = TfidfVectorizer(ngram_range=(1, 5), max_features=80000, tokenizer=(lambda s: s.split()))
X_train = vectorizer.fit_transform(train_sessions)
X_test = vectorizer.transform(test_sessions)
y_train = y.astype('int').values
```

```python
%%time
#new time features 
time_diff =['diff%s' % i for i in range(1, 10)]
def time_to_diff(time):
    time.sort_values(by='time1')
    for j in range(0, 9):
        time[time_diff[j]] = (time[times[j+1]]-time[times[j]]).astype('timedelta64[s]').fillna(0).astype('int')
    time['sum_diff'] = time[time_diff].sum(axis=1)
    return
 
time_diff_train = time_to_diff(train_times)
time_diff_test = time_to_diff(test_times)
```

```python
%%time

#new set features
def prepare_df_set_features(df_csv, df_times, X_sparse):

    times = ['time%s' % i for i in range(1, 11)]
    sites = ['site%s' % i for i in range(1, 11)]   
    
    #year
    year_1 = df_times['time1'].apply(lambda x: int(x.strftime("%Y"))).astype('int').values.reshape(-1, 1)
    day_1 = df_times['time1'].apply(lambda x: int(x.strftime("%d"))).astype('int').values.reshape(-1, 1)
    month_1 = df_times['time1'].apply(lambda x: int(x.strftime("%m"))).astype('int').values.reshape(-1, 1)
    
    #for time10
    year_t10 = df_times['time10'].apply(lambda x: int(x.strftime("%Y"))).astype('int').values.reshape(-1, 1)
    day_t10 = df_times['time10'].apply(lambda x: int(x.strftime("%d"))).astype('int').values.reshape(-1, 1)
    month_t10 = df_times['time10'].apply(lambda x: int(x.strftime("%m"))).astype('int').values.reshape(-1, 1)
   
    #season
    spring = ((month_1 >= 3) & (month_1 < 6)).astype('int').reshape(-1, 1)
    summer = ((month_1 >= 6) & (month_1 < 9)).astype('int').reshape(-1, 1)
    autumn = ((month_1 >= 9) & (month_1 < 12)).astype('int').reshape(-1, 1)
    winter = ((month_1 >= 12) & (month_1 < 3)).astype('int').reshape(-1, 1)
    
    # time year is site1 = site10 
    time1_10 = (year_1 == year_t10).astype('int').reshape(-1, 1)
    
    
    #fortime1
    year1_4 = (month_1 <= 3).astype('int').reshape(-1, 1)
    year2_4 = ((month_1 > 3) & (month_1 <= 6)).astype('int').reshape(-1, 1)
    year3_4 = ((month_1 > 6) & (month_1 <= 9)).astype('int').reshape(-1, 1)
    year4_4 = (month_1 < 9).astype('int').reshape(-1, 1)
    year_2000 = (year_1 >= 2000).astype('int').reshape(-1, 1)
    intercalary = ((year_1 % 4 == 0) | (year_1 % 400 == 0) | (year_1 % 100 == 0)).astype('int').reshape(-1, 1)
    
    #start_hour
    start_hour = df_times['time1'].apply(lambda x: int(x.strftime("%H"))).astype('int').values.reshape(-1, 1)
    start_hour_10 = df_times['time10'].apply(lambda x: int(x.strftime("%H"))).astype('int').values.reshape(-1, 1)
    #start_hour1_min_10
    
    #start_sec
    start_s = df_times['time1'].apply(lambda x: int(x.strftime("%S"))).astype('int').values.reshape(-1, 1)
    start_s_t10 = df_times['time10'].apply(lambda x: int(x.strftime("%S"))).astype('int').values.reshape(-1, 1)

    #day_of_week
    day_of_week = df_times['time1'].apply(lambda x: int(x.strftime("%u"))).astype('int').values.reshape(-1, 1)
    weekday = (day_of_week >= 6).astype('int').reshape(-1, 1)
    workday = (day_of_week < 6).astype('int').reshape(-1, 1)
    day_of_year = df_times['time1'].apply(lambda x: int(x.strftime("%j"))).astype('int').values.reshape(-1, 1)

        #day of months                               
    day_of_months = df_times['time1'].apply(lambda x: int(x.strftime("%d"))).values.reshape(-1, 1)
    day_of_months_1_2 = (day_of_months <= 15).astype('int').reshape(-1, 1)
    day_of_months_2_2 = (day_of_months <= 16).astype('int').reshape(-1, 1)
    
    hour = df_times['time1'].apply(lambda ts: ts.hour).astype('int').values.reshape(-1, 1)
    morning_h = ((hour >= 7) & (hour <= 11)).astype('int').reshape(-1, 1)
    day_h = ((hour >= 12) & (hour <= 18)).astype('int').reshape(-1, 1)
    evening_h = ((hour >= 19) & (hour <= 23)).astype('int').reshape(-1, 1)
    night_h = ((hour >= 0) & (hour <= 6)).astype('int').reshape(-1, 1)

    #session time
       
   # plus_ses = (df_times['sum_diff'].sum(axis=1).values).astype('int').reshape(-1, 1)
    zero_ses = (df_times['sum_diff'].values == 0).astype('int').reshape(-1, 1)
    minus_ses = (df_times['sum_diff'].values < 0).astype('int').reshape(-1, 1)
    long_ses_10 = (abs(df_times['sum_diff'].values) > 10).astype('int').reshape(-1, 1)
    long_ses_100 = (abs(df_times['sum_diff'].values) > 100).astype('int').reshape(-1, 1)
    long_ses_10000 = (abs(df_times['sum_diff'].values) > 10000).astype('int').reshape(-1, 1)
    t_diff2_1  = (abs(df_times['diff1']-df_times['diff2']).values < 10).astype('int').reshape(-1, 1)
    #test_df[sites]
    X = np.concatenate((year_1, day_1, month_1, year_t10, day_t10, month_t10, time1_10,  
                        spring, summer, autumn, winter,  year1_4, year2_4, year3_4, year4_4, intercalary,
                        year_2000, start_hour, start_hour_10, start_s,  start_s_t10,
                        day_of_week, weekday, workday, day_of_year, day_of_months,   day_of_months_1_2,
                        day_of_months_2_2, hour, morning_h, day_h, evening_h, night_h,
                        zero_ses, minus_ses, long_ses_10, long_ses_100, t_diff2_1), axis=1)
#    print("X", X)
    return X


X_fit_test = prepare_df_set_features(test_df, test_times, X_test)
X_fit_test.shape 
```

```python
X_test_all =  csr_matrix(hstack([X_test, X_fit_test]))
```

```python
%%time
X_fit_train = prepare_df_set_features(train_df, train_times,  X_train)
X_train_all = csr_matrix(hstack([X_train, X_fit_train]))
```
